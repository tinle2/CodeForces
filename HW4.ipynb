Problem 1

Hidden weights: WH = [1; 1]

Hidden biases: bH = [-1; -3]

Output weights: WO = [-1, 2]

Output bias: bO = 0.5

Activation: Hard threshold

(a)

Number of hidden units: Nh = 2
Number of output units: No = 1

(b)

Hidden layer linear outputs:

zH1 = x – 1

zH2 = x – 3

(c)

Hidden activations:

uH1 = 1 if x ≥ 1, else 0

uH2 = 1 if x ≥ 3, else 0

(d)

Output layer linear function:
zO = -uH1 + 2·uH2 + 0.5

(e)

Network prediction (ŷ):

ŷ = 1 if zO ≥ 0, else 0

This simplifies to:

x range	ŷ
x < 1	1
1 ≤ x < 3	0
x ≥ 3	1
